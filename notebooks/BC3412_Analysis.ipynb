```
lobster_locations = lobster[['Location_ID', 'Location']].drop_duplicates().sort_values(by='Location_ID').reset_index(drop=True)
print("Unique Location_ID and Location pairs in lobster dataset:")
display(lobster_locations.head(10))
```

```
bay_of_fundy_coords = microplastics[microplastics['Region'] == 'Bay of Fundy'][['Latitude (degree)', 'Longitude (degree)']].mean()
gulf_st_lawrence_coords = microplastics[microplastics['Region'] == 'Gulf of St. Lawrence'][['Latitude (degree)', 'Longitude (degree)']].mean()

print("Mean coordinates for Bay of Fundy:")
print(bay_of_fundy_coords)
print("\nMean coordinates for Gulf of St. Lawrence:")
print(gulf_st_lawrence_coords)
```

```
import numpy as np

# 1. Initialize new columns named Latitude and Longitude in the lobster_locations DataFrame with NaN values.
lobster_locations['Latitude'] = np.nan
lobster_locations['Longitude'] = np.nan

# 2. For the row in lobster_locations where the 'Location' is 'Bay of Fundy', assign the Latitude (degree) and Longitude (degree) from the bay_of_fundy_coords Series.
lobster_locations.loc[lobster_locations['Location'] == 'Bay of Fundy', 'Latitude'] = bay_of_fundy_coords['Latitude (degree)']
lobster_locations.loc[lobster_locations['Location'] == 'Bay of Fundy', 'Longitude'] = bay_of_fundy_coords['Longitude (degree)']

# 3. Identify the lobster locations that are geographically located within the Gulf of St. Lawrence region. Specifically, for 'Malpeque Bay Inside', 'Malpeque Bay Outside', 'Morell/Red Head', 'Borden', 'Egmont Bay', 'Howards Cove', 'Georgetown', and 'Pinette', assign the Latitude (degree) and Longitude (degree) from the gulf_st_lawrence_coords Series.
gulf_st_lawrence_lobster_locations = [
    'Malpeque Bay Inside', 'Malpeque Bay Outside', 'Morell/Red Head', 'Borden',
    'Egmont Bay', 'Howards Cove', 'Georgetown', 'Pinette'
]
lobster_locations.loc[lobster_locations['Location'].isin(gulf_st_lawrence_lobster_locations), 'Latitude'] = gulf_st_lawrence_coords['Latitude (degree)']
lobster_locations.loc[lobster_locations['Location'].isin(gulf_st_lawrence_lobster_locations), 'Longitude'] = gulf_st_lawrence_coords['Longitude (degree)']

# 4. Display the updated lobster_locations DataFrame, showing the assigned coordinates and identifying any remaining unmapped locations.
print("Updated lobster_locations with assigned coordinates:")
display(lobster_locations.head(20))
```

```
from math import radians, sin, cos, sqrt, atan2

def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in kilometers

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Initialize an empty list to store the aggregated microplastics data
aggregated_microplastics_data = []

print("Haversine distance function defined and aggregated_microplastics_data list initialized.")



```

```
radius_km = 200 # Increased radius to capture more data

unified_years = microplastics.loc[microplastics['Year'] != -1, 'Year'].unique()

# Clear the list before re-populating to avoid appending to old data
aggregated_microplastics_data = []

for idx, lobster_loc_row in lobster_locations.iterrows():
    lobster_location_id = lobster_loc_row['Location_ID']
    lobster_location_name = lobster_loc_row['Location']
    lobster_lat = lobster_loc_row['Latitude']
    lobster_lon = lobster_loc_row['Longitude']

    # Skip if lobster location coordinates are NaN (unmapped)
    if pd.isna(lobster_lat) or pd.isna(lobster_lon):
        continue

    for year in unified_years:
        microplastics_yearly = microplastics[(microplastics['Year'] == year) &
                                             (~microplastics['Latitude (degree)'].isna()) &
                                             (~microplastics['Longitude (degree)'].isna())].copy()

        if not microplastics_yearly.empty:
            # Calculate distances
            microplastics_yearly['distance'] = microplastics_yearly.apply(
                lambda row: haversine_distance(lobster_lat, lobster_lon, row['Latitude (degree)'], row['Longitude (degree)']),
                axis=1
            )

            # Filter microplastics samples within the defined radius
            nearby_microplastics = microplastics_yearly[microplastics_yearly['distance'] <= radius_km]

            if not nearby_microplastics.empty:
                avg_concentration = nearby_microplastics['Microplastics Measurement'].mean()
            else:
                avg_concentration = np.nan # No samples within radius
        else:
            avg_concentration = np.nan # No microplastic data for that year

        # Append the aggregated data
        aggregated_microplastics_data.append({
            'Location_ID': lobster_location_id,
            'Location': lobster_location_name,
            'Year': year,
            'Latitude': lobster_lat,
            'Longitude': lobster_lon,
            'Aggregated_Microplastics_Concentration': avg_concentration
        })

# Convert the list of dictionaries into a new pandas DataFrame
aggregated_microplastics_df = pd.DataFrame(aggregated_microplastics_data)

# Display the first few rows of the resulting aggregated microplastics DataFrame
print("Aggregated Microplastics Data:")
display(aggregated_microplastics_df.head())
print(f"Number of non-null microplastic concentrations: {aggregated_microplastics_df['Aggregated_Microplastics_Concentration'].count()}")
```

```
merged_lobster_coords = pd.merge(lobster, lobster_locations[['Location_ID', 'Latitude', 'Longitude']], on='Location_ID', how='left')

final_merged_df = pd.merge(merged_lobster_coords, aggregated_microplastics_df[['Location_ID', 'Year', 'Aggregated_Microplastics_Concentration']],
                           on=['Location_ID', 'Year'], how='left')

print("Final Merged DataFrame Head:")
display(final_merged_df.head())

print("\nFinal Merged DataFrame Info:")
final_merged_df.info()
```

```
regression_df = final_merged_df[['Shellhardness', 'Aggregated_Microplastics_Concentration']].dropna()

print(f"Shape of regression data after dropping NaNs: {regression_df.shape}")
display(regression_df.head())
```

```
import statsmodels.api as sm

X = regression_df['Aggregated_Microplastics_Concentration']
Y = regression_df['Shellhardness']

X = sm.add_constant(X) # Adds a constant term to the predictor

model = sm.OLS(Y, X).fit()

print(model.summary())
```

The regression results above show the statistical relationship between `Aggregated_Microplastics_Concentration` and `Shellhardness`. Key metrics to look at include:

*   **R-squared:** Indicates the proportion of the variance in the dependent variable (`Shellhardness`) that is predictable from the independent variable (`Aggregated_Microplastics_Concentration`).
*   **P-value for 'Aggregated_Microplastics_Concentration':** This tells us if the microplastic concentration is a statistically significant predictor of shell hardness. A p-value less than 0.05 typically indicates significance.
*   **Coeff (coefficient) for 'Aggregated_Microplastics_Concentration':** This is the estimated change in `Shellhardness` for every one-unit increase in `Aggregated_Microplastics_Concentration`.
